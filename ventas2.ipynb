```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd

# Define el alcance y carga tus credenciales
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('credentials.json', scope)

# Autenticación con gspread
client = gspread.authorize(creds)

# Abrir la hoja de cálculo por nombre
sheet = client.open("ventas_2").sheet1


#convertir datos a pandas
dataframe=pd.DataFrame(sheet.get_all_records())

dataframe.head()
```

```python
cantidad_vendida=dataframe['Cantidad Vendida']
precio_unitario=dataframe['Precio Unitario']
dataframe["Total de ventas"]=cantidad_vendida*precio_unitario
dataframe.head()
```

```python
ventas_por_producto=dataframe.groupby('Producto')['Total de ventas'].sum()
ventas=dataframe['Total de ventas'].sum()
print(ventas_por_producto)
print(f"total ventas: {ventas}")
```
